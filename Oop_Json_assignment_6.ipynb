```python
# Assignment 1

''' 1. Create a JSON file (employee.json) containing employee information of minimum 5 employees.
Each employee information consists of Name, DOB, Height, City, State. Write a python program that
reads this information from the JSON file and saves the information into a list of objects of Employee 
class. Finally print the list of the Employee objects.'''

import json

# Create employee information of 5 employees
employee_data =[
    {
        "Name": "Avishek Kumar",
        "DOB": "1990-01-01",
        "Height": 175,
        "City": "Noida",
        "State": "UP"
    }, 
    {
        "Name": "Vedant Prasad",
        "DOB": "1985-05-12",
        "Height": 162,
        "City": "Bengalore",
        "State": "Karnataka"
    },
    {
        "Name": "Raghav Chauhan",
        "DOB": "1992-09-20",
        "Height": 180,
        "City": "Pune",
        "State": "Maharashtra"
    },
    {
        "Name": "Anubhav Raj",
        "DOB": "1993-03-28",
        "Height": 168,
        "City": "Bhubaneswar",
        "State": "Odisha"
    },
    {
        "Name": "Pratik Gupta",
        "DOB": "1988-11-15",
        "Height": 170,
        "City": "Patna",
        "State": "Bihar"
    }
]

with open('files/employee.json', 'w') as file:
    json.dump(employee_data, file)

print("JSON file created successfully!")

```

```python

import json

class Employee:
    def __init__(self, name, dob, height, city, state):
        self.name = name
        self.dob = dob
        self.height = height
        self.city = city
        self.state = state

# Read employee information from the JSON file
with open('files/employee.json', 'r') as file:
    employee_data = json.load(file)

employee_list = []

# Create Employee objects and add them to the list
for employee in employee_data:
    name = employee['Name']
    dob = employee['DOB']
    height = employee['Height']
    city = employee['City']
    state = employee['State']

    emp = Employee(name, dob, height, city, state)
    employee_list.append(emp)


# Print the list of Employee objects
for emp in employee_list:
    print("Name:", emp.name)
    print("DOB:", emp.dob)
    print("Height:", emp.height)
    print("City:", emp.city)
    print("State:", emp.state)
    print()
```

```python
## Assignment 2

''' 1. Create a class named â€˜Dogâ€™. It should have a constructor which accepts its name, age and coat color. You must perform the following operations:

ðŸ”´ a. It should have a function â€˜description()â€™ which prints the name and age of the dog.
ðŸ”´ b. It should have a function â€˜get_info()â€™ which prints the coat color of the dog.
ðŸ”´ c. Create child classes â€˜JackRussellTerrierâ€™ and â€˜Bulldogâ€™ which is inherited from the class â€˜Dogâ€™. It should have at least two methods of its own.
ðŸ”´ d. Create objects and implement the above functionalities.'''


# create class

class Dog:
    def __init__(self, name, age, coat_color):
        self.name = name
        self.age = age
        self.coat_color = coat_color

    def description(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

    def get_info(self):
        print(f"Coat Color: {self.coat_color}")
        

# create child class â€˜JackRussellTerrierâ€™
class JackRussellTerrier(Dog):
    def __init__(self, name, age, coat_color):
        super().__init__(name, age, coat_color)

    def jump(self):
        print(f"{self.name} is jumping!")

    def hunt(self):
        print(f"{self.name} is hunting!")
        

# create child class â€˜Bulldogâ€™ 
class Bulldog(Dog):
    def __init__(self, name, age, coat_color):
        super().__init__(name, age, coat_color)

    def snore(self):
        print(f"{self.name} is snoring!")

    def guard(self):
        print(f"{self.name} is guarding!")


# Create objects 
dog1 = Dog("Jimmy", 3, "Brown")
dog1.description()
dog1.get_info()

dog2 = JackRussellTerrier("Jaffy", 5, "White and Brown")
dog2.description()
dog2.get_info()
dog2.jump()
dog2.hunt()

dog3 = Bulldog("Rocky", 4, "Black")
dog3.description()
dog3.get_info()
dog3.snore()
dog3.guard()
```
